# –¶–∏–∫–ª—ã, –∏—Ç–µ—Ä–∞—Ç–æ—Ä—ã –∏ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä—ã —Å–ø–∏—Å–∫–æ–≤

## 5.1 –û–ø–µ—Ä–∞—Ç–æ—Ä —Ü–∏–∫–ª–∞ while

–®–∞–≥ 4. –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º—ã –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã—Ö —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª, –ø—Ä–∏—á–µ–º, n < m. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ—á–∏—Ç–∞—Ç—å —ç—Ç–∏ —á–∏—Å–ª–∞ –∏ –≤—ã–≤–µ—Å—Ç–∏ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª –∫–≤–∞–¥—Ä–∞—Ç—ã —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ [n; m]. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
n,m = map(int, input().split())
lst=[]
while n <= m:
    lst.append(n**2)
    n += 1
print(*lst)
```

–®–∞–≥ 5.  –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º—ã –ø–æ–¥–∞–µ—Ç—Å—è –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ: —Å—Ç–æ–∏–º–æ—Å—Ç—å –æ–¥–Ω–æ–π –∫–Ω–∏–≥–∏ x —Ä—É–±–ª–µ–π. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ—á–∏—Ç–∞—Ç—å —ç—Ç–æ —á–∏—Å–ª–æ –∏ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª —Å—Ç–æ–∏–º–æ—Å—Ç–∏ 2, 3, ... 10-—Ç–∏ —Ç–∞–∫–∏—Ö –∫–Ω–∏–≥ —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ –¥–µ—Å—è—Ç—ã—Ö. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
p = float(input())
n = 2
while n <= 10:
    print(round(n*p,1), end=' ')
    n +=1
```

–®–∞–≥ 6. –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º—ã –ø–æ–¥–∞–µ—Ç—Å—è —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —ç—Ç–æ —á–∏—Å–ª–æ, –∞ –∑–∞—Ç–µ–º, –≤—ã—á–∏—Å–ª–∏—Ç–µ –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω —Å–ª–µ–¥—É—é—â—É—é —Å—É–º–º—É —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ —Ç—ã—Å—è—á–Ω—ã—Ö (—Ç—Ä–∏ –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π):

ùëÜ

ùëõ

–ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
n = int(input())
s = 0
i = 1
while i <= n:
    s += 1/i
    i += 1
print(round(s, 3))
```

–®–∞–≥ 7. –ù–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤ –∫–æ—Ç–æ—Ä–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Ü–∏–∫–ª–∞ (while) –¥–æ–ª–∂–µ–Ω –≤–≤–æ–¥–∏—Ç—å —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ. –¶–∏–∫–ª –¥–æ–ª–∂–µ–Ω –ø—Ä–æ–¥–æ–ª–∂–∞—Ç—å—Å—è, –ø–æ–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–µ –≤–≤–µ–¥–µ—Ç —á–∏—Å–ª–æ 0. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã—á–∏—Å–ª–∏—Ç—å —Å—É–º–º—É –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –≤ —Ü–∏–∫–ª–µ —á–∏—Å–µ–ª –∏ –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç (—Å—É–º–º—É) –Ω–∞ —ç–∫—Ä–∞–Ω. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
d = 1
s=0
while d != 0:
    d = int(input())
    s += d
print(s)
```

–®–∞–≥ 8. –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ (—Å–ª–∞–≥). –ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —ç—Ç—É —Å—Ç—Ä–æ–∫—É –∏ –∑–∞–º–µ–Ω–∏—Ç–µ –≤ –Ω–µ–π –≤—Å–µ –ø–æ–¥—Ä—è–¥ –∏–¥—É—â–∏–µ –¥–µ—Ñ–∏—Å—ã (--, ---, ---- –∏ —Ç.–¥.) –Ω–∞ –æ–¥–∏–Ω–∞—Ä–Ω—ã–µ (-). –†–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Å—Ç—Ä–æ–∫–∏ –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
s = input()
while '--' in s:
    s = s.replace('--','-')
print(s)
```

–®–∞–≥ 9. –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ (—Ç–æ –µ—Å—Ç—å, —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ) –æ—Ç —Ç—Ä–µ—Ö–∑–Ω–∞—á–Ω–æ–≥–æ –∏ –±–æ–ª–µ–µ. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ—á–∏—Ç–∞—Ç—å —ç—Ç–æ —á–∏—Å–ª–æ –∏ –Ω–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤—Å–µ—Ö –µ–≥–æ —Ü–∏—Ñ—Ä. –†–µ–∑—É–ª—å—Ç–∞—Ç (–ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ) –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
a = int(input())
p=1
while a:
    p *= a % 10
    a //= 10
print(p)
```

–®–∞–≥ 10. –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –§–∏–±–æ–Ω–∞—á—á–∏ –æ–±—Ä–∞–∑—É–µ—Ç—Å—è —Ç–∞–∫: –ø–µ—Ä–≤—ã–µ –¥–≤–∞ —á–∏—Å–ª–∞ —Ä–∞–≤–Ω—ã 1 –∏ 1, –∞ –∫–∞–∂–¥–æ–µ –ø–æ—Å–ª–µ–¥—É—é—â–µ–µ —Ä–∞–≤–Ω–æ —Å—É–º–º–µ –¥–≤—É—Ö –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö. –í –∏—Ç–æ–≥–µ, –ø–æ–ª—É—á–∞–µ–º —Ç–∞–∫—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª: 1, 1, 2, 3, 5, 8, 13, ...

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n (n >= 2). –ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —ç—Ç–æ —á–∏—Å–ª–æ –∏ —Å—Ñ–æ—Ä–º–∏—Ä—É–π—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –§–∏–±–æ–Ω–∞—á—á–∏ –¥–ª–∏–Ω–æ–π n. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–∏ n = 4 –ø–æ–ª—É—á–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å:

1 1 2 3

–†–µ–∑—É–ª—å—Ç–∞—Ç –æ—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã—Ö —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
n = int(input())
a,b=1,1
count=1
while count <= n:
    print(a, end=' ')
    count += 1
    a,b = b, a+b
```

–®–∞–≥ 11. –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∞—Å–æ–≤). –ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —ç—Ç–æ —á–∏—Å–ª–æ –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π n.

–ü—É—Å—Ç—å –æ–¥–Ω–æ–∫–ª–µ—Ç–æ—á–Ω–∞—è –∞–º–µ–±–∞ –∫–∞–∂–¥—ã–µ 3 —á–∞—Å–∞ –¥–µ–ª–∏—Ç—Å—è –Ω–∞ 2 –∫–ª–µ—Ç–∫–∏. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, —Å–∫–æ–ª—å–∫–æ –∫–ª–µ—Ç–æ–∫ –±—É–¥–µ—Ç —á–µ—Ä–µ–∑ n —á–∞—Å–æ–≤. –°—á–∏—Ç–∞—Ç—å, —á—Ç–æ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –±—ã–ª–∞ –æ–¥–Ω–∞ –∞–º–µ–±–∞.

–†–µ–∑—É–ª—å—Ç–∞—Ç (–∏—Ç–æ–≥–æ–≤–æ–µ —á–∏—Å–ª–æ –∫–ª–µ—Ç–æ–∫) –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω. –ó–∞–¥–∞—á—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–µ—à–∏—Ç—å —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ü–∏–∫–ª–∞ while.

```python
n = int(input())
c=1
while n//3 > 0:
    c *= 2
    n -= 3
print(c)
```

–®–∞–≥ 12. –ì—Ä–∞–∂–¥–∞–Ω–∏–Ω 1 —è–Ω–≤–∞—Ä—è –æ—Ç–∫—Ä—ã–ª —Å—á–µ—Ç –≤ –±–∞–Ω–∫–µ, –≤–ª–æ–∂–∏–≤ 1000 —Ä—É–±. –ö–∞–∂–¥—ã–π –≥–æ–¥ —Ä–∞–∑–º–µ—Ä –≤–∫–ª–∞–¥–∞ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –Ω–∞ 5% –æ—Ç –∏–º–µ—é—â–µ–π—Å—è —Å—É–º–º—ã. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Å—É–º–º—É –≤–∫–ª–∞–¥–∞ —á–µ—Ä–µ–∑ n –ª–µ—Ç (n - —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, —á–∏—Ç–∞–µ–º–æ–µ –∏–∑ –≤—Ö–æ–¥–Ω–æ–≥–æ –ø–æ—Ç–æ–∫–∞). –†–µ–∑—É–ª—å—Ç–∞—Ç (—Å—É–º–º—É –≤–∫–ª–∞–¥–∞) –æ–∫—Ä—É–≥–ª–∏—Ç—å –¥–æ —Å–æ—Ç—ã—Ö –∏ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while

```python
n = int(input())
sum = 1000
year=1
while year <= n:
    sum *= 1.05
    year += 1
print(round(sum, 2))
```

–®–∞–≥ 13. –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ n –∏ m, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã—Ö –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª, –ø—Ä–∏—á–µ–º n < m. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ—á–∏—Ç–∞—Ç—å —ç—Ç–∏ —á–∏—Å–ª–∞ –∏ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å –≤—Å–µ –Ω–µ—á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ [n, m]. –ó–∞–¥–∞—á—É —Å–ª–µ–¥—É–µ—Ç —Ä–µ—à–∏—Ç—å –±–µ–∑ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è —É—Å–ª–æ–≤–Ω–æ–≥–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞. –†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏ —á–∏—Å–µ–ª, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã—Ö —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
n,m = map(int,input().split())
i=n+1
while i<m:
    print(i, end=' ')
    i += 2
```

–®–∞–≥ 14. –°–æ—Å—Ç–∞–≤–∏—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É –ø–æ–∏—Å–∫–∞ –≤—Å–µ—Ö —Ç—Ä–µ—Ö–∑–Ω–∞—á–Ω—ã—Ö —á–∏—Å–µ–ª, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–∏ –¥–µ–ª–µ–Ω–∏–∏ –Ω–∞ 47 –¥–∞—é—Ç –≤ –æ—Å—Ç–∞—Ç–∫–µ 43 –∏ –∫—Ä–∞—Ç–Ω—ã 3. –í—ã–≤–µ—Å—Ç–∏ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è. –ü—Ä–æ–≥—Ä–∞–º–º—É —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ while.

```python
n=100
while n < 1000:
    if n % 3 == 0 and n % 47 == 43:
        print(n, end=' ')
    n += 1
```

## 5.2 –û–ø–µ—Ä–∞—Ç–æ—Ä—ã break, continue –∏ else

–®–∞–≥ 4. –û–±—ä—è–≤–∏—Ç–µ –≤ –ø—Ä–æ–≥—Ä–∞–º–º–µ —Å–ª–µ–¥—É—é—â–∏–π –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π —Å–ø–∏—Å–æ–∫ –¥–ª–∏–Ω–æ–π 10 —ç–ª–µ–º–µ–Ω—Ç–æ–≤, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –Ω—É–ª–µ–π:

p = [0] * 10
–ù–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Ü–∏–∫–ª–∞ –∑–∞–ø—Ä–∞—à–∏–≤–∞–π—Ç–µ —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤–≤–æ–¥ —Ü–µ–ª–æ–≥–æ —á–∏—Å–ª–∞ - –∏–Ω–¥–µ–∫—Å —ç–ª–µ–º–µ–Ω—Ç–∞ —Å–ø–∏—Å–∫–∞ p, –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ 1, –µ—Å–ª–∏ –µ–µ —Ç–∞–º –µ—â–µ –Ω–µ—Ç. –ï—Å–ª–∏ –∂–µ 1 —É–∂–µ –ø—Ä–æ—Å—Ç–∞–≤–ª–µ–Ω–∞ –ø–æ —Ç–µ–∫—É—â–µ–º—É –∏–Ω–¥–µ–∫—Å—É, —Ç–æ —Å–ø–∏—Å–æ–∫ –Ω–µ –º–µ–Ω—è—Ç—å. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–∞—Å—Å—Ç–∞–≤–∏—Ç—å —Ç–∞–∫ –ø—è—Ç—å –µ–¥–∏–Ω–∏—Ü –≤ —Å–ø–∏—Å–∫–µ p. –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ —Ü–∏–∫–ª –ø—Ä–µ—Ä—ã–≤–∞–µ—Ç—Å—è.
